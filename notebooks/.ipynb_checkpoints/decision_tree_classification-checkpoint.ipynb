```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier ,plot_tree
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/media/utkarsh/Working/mlp/workspace/student_grade_predictor/data/main_data.csv')
data.head()
```

```python
x=data.drop('final-grade',axis=1)
y=data['final-grade']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
dtc = DecisionTreeClassifier(criterion='gini', max_depth=5, random_state=42)
dtc.fit(X_train, y_train)
y_pred = dtc.predict(X_test)
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nAccuracy Score:", accuracy_score(y_test, y_pred))
```

```python
plt.figure(figsize=(15, 10))
plot_tree(dtc, feature_names=x.columns, filled=True)
plt.show()
```
